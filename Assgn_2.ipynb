# What is the distribition of contries in top 100 universities?

## What this means:
You're being asked to analyze how the top 100 universities are spread across different countries.
In plain English:
• Which countries have universities in the top 100?
• How many universities does each country have in the top 100?
• Is the list dominated by a few countries (e.g., US, UK)?
• Are some countries represented only once?
Step-by-step Approach in pandas:
Assume:
• Your DataFrame is df
• The ranking column is 'University Rank' (numeric or already converted)
• The country column is 'Location' (adjust based on your dataset)

```python
# top 100 universities acc to teh rank 

df['University Rank'] = pd. to_numeric(df['University Rank'], errors= 'coerce')
#  This will turn strings like '1', '2', etc., into integers.
#  If there are any non-numeric values (like "—", n/a)

top_100_univ = df[df['University Rank'] <=100]

top_100_univ
```

#  What is the average number of students in top 10 universities?

```python
top_10_univ = df[df['University Rank'] <=10]
top_10_univ
```

```python

# first converting col no. od students string ttype to numeric type 
# top_10_univ['No of student'] = pd. to_numeric(top_10_univ['No of student'], errors= 'coerce')


top_10_univ = df[df['University Rank'] <= 10].copy()

# cleaning col before converting 
top_10_univ['No of student'] = top_10_univ['No of student'].astype(str).str.replace(',', '').str.strip()

# now converting into numeric
top_10_univ['No of student'] = pd.to_numeric(top_10_univ['No of student'], errors='coerce')

top_10_univ
```

```python
# counting avg students in top 10 universities
avg_stu_10_univ= top_10_univ['No of student'].mean()
print('average number of students in top 10 universities?',avg_stu_10_univ)
```

# Which are the top 10 universities with more than 50% international student?

```python
# Clean 'International Student' column
df['International Student'] = df['International Student'].str.rstrip('%')
df['International Student'] = pd.to_numeric(df['International Student'], errors='coerce')

# Filter for > 50%
filtered_univs = df[df['International Student'] > 50]

# Sort by rank and take top 10
top_10_int_students = filtered_univs.sort_values('University Rank').head(10)

# Show relevant columns
print(top_10_int_students[['University Rank', 'Name of University', 'International Student']])
```

# Which are the top 10 universities with a predominantly female presence?

```python
# Step 1: Parse the ratio
ratios = df['Female:Male Ratio'].str.split(':', expand=True)
df['Female_ratio'] = pd.to_numeric(ratios[0].str.strip(), errors='coerce')
df['Male_ratio'] = pd.to_numeric(ratios[1].str.strip(), errors='coerce')

# Step 2: Filter universities with more females than males
female_predominant = df[df['Female_ratio'] > df['Male_ratio']]

# Step 3: Get top 10 by rank
top_10_female = female_predominant.sort_values('University Rank').head(10)

# Step 4: Show results
print(top_10_female[['University Rank', 'Name of University', 'Female:Male Ratio']])
```
